## Desarrollo

### Data Wrangling

Lo primero que necesitamos es hacer un proceso de *Data Wrangling* para tener nuestros datos lo más limpios posible. Para ello, primero cargamos el conjunto de datos proporcionado:

````python
df = pd.read_excel('cia_países.xlsx')
df
````

Ahora queremos identificar el tipo de datos en cada columna, en este caso, todas las columnas a excepción de **Name** y **Continent** deberían ser valores numéricos, ya sea **int** o **float**.

````python
df.dtypes
````

Notemos que casi todos los datos tienen su tipo de dato correspondiente. Las columnas que pueden llegar a presentar problemas son las de **Area (km²)** y la de **Population**, ya que están marcadas como **object**. Veamos cómo se comportan estos datos, empecemos con la columna de población:

````python
#for i in df['Population']:
#    print(i)
````

Notemos que hay datos que no tienen valores enteros, algunos tienen valor "no", otros "uninhabited" y uno que es flotante con 67.081. Veamos a qué filas les corresponde:

````python
df[df['Population'] == "uninhabited"]
````

````python
df[df['Population'] == "no"]
````

````python
df[df['Population'] == 67.081]
````

Veamos que los lugares con valores de "no" o de "uninhabited" son filas mayormente con valores nulos. Algunas (si es que no todas) no podrían considerarse como países, ya que son islas, territorios océanicos o zonas inhabitables. Por esta razón vamos a quitar de nuestro conjunto cualquier fila que presente alguna de estas características.

Por otro lado, el valor de **Population** con 67.081 le pertenece al Reino Unido. Vamos a asumir que su dato está expresado en unidades de millón y lo vamos a multiplicar por 1,000,000.

Aplicando los cambios a nuestro *DataFrame* tenemos:

````python
counter = 0
for i in df['Population']:
    #print(i)
    if i == 67.081:
        #print(i)
        df.loc[df.index[counter], 'Population'] = int(i*1000000)
    if(i in ["no","uninhabited"]):
        df.drop(counter, inplace = True)
    counter += 1
df['Population'] = df['Population'].astype(int)
df['Population']
````

Ya tenemos arreglada la columna de **Poblacion**, falta revisar la de **Area (km²)**. Veamos sus datos:

````python
#for i in df['Area (km²)']:
#    print(i)
````

Notemos que algunos valores son **string** que tienen unidades expresadas en millones. Otros tienen solo un punto, lo cual asumiremos también indica que están en unidades de millones. Vamos a transformar todos los valores en **string**, quitar todas las comas y palabras de más, al final expresaremos todo de manera numérica de nuevo:

````python
df['Area (km²)'] = df['Area (km²)'].astype(str)
df['Area (km²)'] = df['Area (km²)'].str.split().str[0]
df['Area (km²)'] = df['Area (km²)'].str.replace(',','')
counter = 0
for i in df['Area (km²)']:
    if '.' in i:
        #print(i)
        df.loc[df.index[counter], 'Area (km²)'] = int(float(i)*1000000)
    counter += 1
df['Area (km²)'] = df['Area (km²)'].astype(int)
df['Area (km²)']
````

Ya tenemos todos los tipos de datos que queremos en cada columna:

````python
df.dtypes
````

Por si acaso, algo más que haremos será identificar datos atípicos. Recordemos que identificamos varios de estos al evaluar la columna de **Population**, ya que no tenían habitantes. Algo que podría también suceder es que tengamos entre nuestros datos a organizaciones de naciones enteras que no sean realmente un país. Una forma posible de poder identificarlos sería viendo las filas con las cantidades más grandes de población. Veamos las primeras 25:

````python
df.sort_values(by=['Population'], ignore_index=True, ascending = False).head(25)
````

Notemos que sí tenemos dos datos que no son países en nuestro *DataFrame*. Una fila es la Unión Europea y otra es el mundo entero (curiosamente hay un error porque la población del mundo es menor que la de otros países). Vamos a eliminar estas filas de nuestro *DataFrame*, no deberían pertenecer a nuestros datos:

````python
df = df[df.Name != 'World']
df = df[df.Name != 'European Union']
````

Recordemos que para aplicar modelos de aprendizaje a nuestro conjunto de datos es importante no tener valores nulos. Veamos cuál es el porcentaje de valores nulos de cada columna:

````python
for c in df.columns:
    miss = df[c].isna().sum()/len(df)
    if miss > 0:
        print("{} tiene el {} de valores NaN".format(c, miss))
    else:
        print("{} NO tiene valores faltantes".format(c))
````

Notemos que la mayoría de nuestras columnas tiene un porcentaje bajo de valores NaN, la única con un porcentaje de valores nulos mucho mayor es el de **Militar Expenditures**. De todas formas este último, no es un porcentaje tan alto como para tirar toda la columna. Lo que haremos será reemplazar todos los valores NaN por la mediana de cada columna.

````python
df = df.fillna(df.median())
df
````

Desafortunadamente, hacer este tipo de acciones puede causar inconsistencias en algunos de los datos. Por ejemplo, la población de un país puede tener un valor de 10, pero si la mediana de la columna de número de usuarios de Internet llega a ser algo como 100, entonces esto no tendría sentido. Porque Esto tendríamos más usuarios de Internet que población si este último valor era originalmente NaN.

De momento esto no nos importa mucho porque para evaluar nuestros modelos de entrenamiento vamos a considerar solo algunas de las columnas que sí pueden llegar a ser relevantes para entrenar el modelo. Por ejemplo, no voy a considerar la columna de número de usuarios de Internet si ya tengo la del porcentaje.

Lo que sigue es la transformación en dato categórico de **GDP (USA)**. Las reglas son las siguientes:
* GDP $<$ 50: dato atípico (se elimina)
* 50 $\leq$ GDP $\leq$ 5,000: ingreso bajo
* 5,000 $<$ GDP $\leq$ 25,000: ingreso medio
* GDP $>$ 25,000: ingreso alto

Apliquemos esto a la columna del *DataFrame*:

````python
def to_categoric(x):
    if(x < 50):
        return np.nan
    elif (50 <= x and x <= 5000):
        return 'Ingreso bajo'
    elif (5000 < x and x <= 25000):
        return 'Ingreso medio'
    elif (x > 25000):
        return 'Ingreso alto'
    else:
        return np.nan
df['GDP (USD)'] = df['GDP (USD)'].apply(to_categoric)

# Eliminar filas con valores atípicos
df = df[df['GDP (USD)'].notna()]

df
````

Ya hicimos todo el proceso que considero necesario para tener nuestros datos en la mejor condición posible y así aplicarlos a distintos modelos de entrenamiento. Pasemos a la siguiente etapa.

### Aplicar modelos de entrenamiento

#### Normalización y codificación de datos

Recordemos que para poder aplicar nuestros modelos de aprendizaje es recomendable normalizar y codificar los datos que vayamos a utilizar. Para ello, lo primero que haremos será derterminar cuáles columnas pueden resultar útiles al momento de aplicar el modelo de entrenamiento. En este caso vamos a considerar todas a excepción de: **Name** e **Internet Users**. La primera no la vamos a aplicar porque el nombre no debería ser algo que influencia en el GDP de un país, además es distinto para cada fila. La segunda columna mencionada la vamos a descartar porque creo quye ya tenemos suficiente información de este campo con el porcentaje de Internet.

Adicionalmente, la razón por la que mantengo una columna como la de **Continent**, es que a diferencia de **Name**, los valores se repiten a lo largo de varias filas; es un dato categórico. Puede que el continente al que pertenezca determinado país influencie en menor o mayor medida el GDP de algún país.

Lo primero que hacemos es quitar las columnas que no íbamos a usar para hacer los modelos:

````python
df_data = df.drop(['Name', 'Internet Users'], axis = 1)
````

Lo que sigue es normalizar y codificar el resto de las columnas que vamos a estar utilizando. Lo primero sería codificar nuestras variables categóricas. Dos formas de hacer esto es con ```LabelEncoder``` y con ```OneHotEncoder```, funciones ofrecidas por **sklearn**. La primera nos permite redifinir todas las clases de dato categórico como valores enteros numéricos. La segunda separa en columnas cada categoría e indica con valores de 1 o 0 si pertenece o no a la clase de la columna correspondiente.

Lo conveniente es usar ```LabelEncoder``` para la columna de **GDP (USD)**, cuyos valores son los que vamos a estar prediciendo. Para el caso de **Continent** podríamos hacer lo mismo, pero como no son parte del vector de valores que queremos predecir, vamos a probar usar un ```OneHotEncoder``` en ellos. Para hacer eso, realizamos lo siguiente:

````python
from sklearn import preprocessing

le = preprocessing.LabelEncoder()

# OHE
df_data = df_data.join(pd.get_dummies(df_data['Continent']))
df_data = df_data.drop('Continent', axis = 1)

# Label Encoding
df_data['GDP (USD)'] = le.fit_transform(df_data['GDP (USD)'])
````

Ahora que ya están las columnas categóricas en un formato ideal, lo que sigue es normalizar las columnas con datos numéricos. Para ello podemos aplicar ```MinMaxScaler```, la cual es una función de **sklearn** que nos permite normalizar datos de todas las columnas que indiquemos.

````python
scaler = preprocessing.MinMaxScaler()

columns = ['Area (km²)',
        'Population',
        'Unemployment Rate',
        'Taxes (% of GDP)',
        'Exchange Rate (USD)',
        'External Debt (USD)',
        'Internet Users Percentage of Population',
        'Airports',
        'Roadways (km)',
        'Militar Expenditures (% of GDP)']

df_data[columns] = scaler.fit_transform(df_data[columns])
````

#### Separación de columna a predecir

Ahora debemos separar la columna cuyos valores queremos predecir del resto del *DataFrame*. De esta manera podemos indicarle explícitamente a los modelos de entrenamiento qué es lo que va a predecir y sobre qué datos apoyarse para ello. Separamos en un arreglo *y*, la columna objetivo, **GDP (USD)**, y en una matriz *X* el resto de las columnas del *DataFrame* que vamos a usar en el modelo.

````python
X = df_data.drop(['GDP (USD)'], axis = 1).values
y = df_data['GDP (USD)'].values
````

````python
#print(f'X = {X}\n\ny = {y}')
````

#### Definición de los modelos de aprendizaje automático

A continuación tenemos que definir los modelos de aprendizaje automático supervisado que vamos a utilizar para este conjunto de datos. Lo que hay que tomar en cuenta para esta selección es que estamos trabajando con problemas de clasificación de múltiples categorías, así que hay que considerar únicamente modelos que hagan este tipo de trabajo. Hay varios de estos, por ejemplo:
* Naive Bayes
* Regresión Logística
* Máquinas de Soporte Vectorial
* K-Nearest Neighbors

Para este proyecto vamos a utilizar los 3 últimos. La librería de **sklearn** nos permite aplicar todos estos modelos de entrenamiento, así que definamos los tres clasificadores de una vez:

**Nota**: en el modelo de ```SVC``` podemos especificar un kernel. En este caso estuve experimentando con cuatro de los que ofrece: *linear*, *rbf*, *poly* y *sigmoid*. Como *linear* fue el que me arrojó mejores resultados en esta ocasión, decidí quedarme con ese. Si usaba algo como *rbf* obtenía peores resultados y esto provocaba que la regresión logística fuera el modelo ganador al final del proyecto. Como en este caso *lineal* era el de mejor desempeño, opté por quedarme con ese y resultó en tener mejores puntajes que la regresión logística.

````python
from sklearn.linear_model import LogisticRegression
from sklearn.svm import SVC
from sklearn.neighbors import KNeighborsClassifier
#from sklearn.linear_model import SGDClassifier  #Esto es de pruebas con las que quise experimentar por separado
#from sklearn.naive_bayes import GaussianNB

lr = LogisticRegression(multi_class='ovr')
svc = SVC(kernel = 'linear', decision_function_shape = 'ovo') # Probé kernels 'linear', 'rbf', 'poly' y 'sigmoid', 
                                                            #el que mejor resultados dio en este caso fue linear 
knn = KNeighborsClassifier()
#nb = GaussianNB()
#sgd_clf.fit(X_norm, y_norm)
#cross_val_score(sgd_clf,X_norm, y_norm, cv = 5, scoring = 'accuracy')
````

#### Midiendo desempeño con validación cruzada

Ya que tenemos definidos nuestros tres clasificadores, vamos a evaluar el desempeño de cada uno sobre el conjunto de datos con validación cruzada. Para ello utilizaremos las cuatro métricas distintas que nos pide el problema:
* Exactitud (accuracy)
* Precisión
* Sensibilidad (recall)
* $F_1$

Definamos una función que se encargue de imprimir dado un modelo, el resultado de la validación cruzada para las distintas métricas. Para cada resultado de la validación cruzada, especificará el arreglo obtenido y su promedio. Finalmente dará un puntaje general que será el promedio de los promedios obtenido de cada métrica, este puntaje general lo definiremos como el desempeño total del clasificador.

Para hacer la validación cruzada vamos a aplicar la función de  ```cross_val_score```, a la cual le especificaremos $k=5$, el clasificador, nuestra matriz de datos, el vector de etiquetas y finalmente un **scoring**. Este **scoring** define la métrica que vamos a estar utilizando. Vamos a estar utilizando las métricas que ya mencionamos; pero hay que tener cuidado en cómo definimos a las de $F_1$, precisión y sensibilidad porque por defecto están pensadas para clasificación binaria. La solución a esto se encuentra en que le especifiquemos el *average* que va a considerar, las opciones que nos ofrece **sklearn** son *macro*, *micro* y *weighted*.

Acorde al [código](https://github.com/scikit-learn/scikit-learn/blob/7b136e9/sklearn/metrics/classification.py#L620) de **sklearn**, tenemos que *micro* calcula las métricas globales de todo el conjunto de datos, por falsos negativos, falsos positivos y positivos verdaderos; *macro* calcula las métricas de cada etiqueta y obtiene el promedio sin considerar la ponderación, esto quiere decir, que no va a tomar en cuenta el desbalance que hay en las proporciones de las tres distintas etiquetas que tenemos para **GDP (USD)**; finalmente está *weighted* que también calcula las métricas de cada etiqueta y además encuentra un promedio ponderado, considerando así las proporciones de las etiquetas.

En este caso, pienso que no hay mucho problema en usar cualquiera de las tres. Pero por si acaso, podemos ver las proporciones que tenemos de cada etiqueta:

````python
df_data['GDP (USD)'].value_counts(normalize = True)
````

Vemos que las etiquetas sí están un tanto desbalanceadas. Más del 40% le corresponden a una sola, la más baja tiene un porcentaje de casi la mitad de la más grande y la otra se encuentra como en el medio de las dos. Dado esto, lo que haremos será utilizar un *average* que considere estas proporciones, *weighted* puede ser una opción aceptable en este caso.

Dicho todo lo anterior ahora sí definimos la función que mencionamos:

````python
from sklearn.model_selection import cross_val_score

def cross_val_print(model, matrix, vector, k):
    ac = cross_val_score(model, matrix, vector, cv = k, scoring="accuracy") # Accuracy 
    pr = cross_val_score(model, matrix, vector, cv = k, scoring="precision_weighted") # Precision
    re = cross_val_score(model, matrix, vector, cv = k, scoring="recall_weighted") # Recall
    f1 = cross_val_score(model, matrix, vector, cv = k, scoring="f1_weighted") # F1
    print("Puntajes de métricas en validación cruzada:")
    print(f"Accuracy:\n\tVector = {ac}\n\tPromedio = {ac.mean()}")
    print(f"Precision:\n\tVector = {pr}\n\tPromedio = {pr.mean()}")
    print(f"Recall:\n\tVector = {re}\n\tPromedio = {re.mean()}")
    print(f"F1:\n\tVector = {f1}\n\tPromedio = {f1.mean()}")
    print(f"\nPromedio global (promedio de los promedios) = {(ac.mean() + pr.mean() + re.mean() + f1.mean())/4}")
````

Ahora imprimimos para cada clasificador su desempeño acorde a cada métrica y su "promedio global".

````python
aux = "/////////////////////////////////////////////////////////////////////////"

print("REGRESIÓN LOGÍSTICA\n")
cross_val_print(lr, X, y, 5)
print("\n" + aux)
print("\nMÁQUINA DE SOPORTE VECTORIAL\n")
cross_val_print(svc, X, y, 5)
print("\n" + aux)
print("\nkNN\n")
cross_val_print(knn, X, y, 5)
print("\n\n")
````

Ya tenemos un registro del desempeño de cada modelo en distintas métricas usando la validación cruzada. El último paso es determinar al modelo ganador en nuestro conjunto de datos.

#### Determinación del modelo ganador

Para determinar el modelo ganador basta con observar los distintos puntajes que obtuvimos para cada modelo. Si nos fijamos, los tres modelos tienen puntajes muy parecidos entre sí para todas sus métricas y puntaje del desempeño general. Notemos cómo en todas las métricas, el modelo de **SVC** que definimos es el que obtuvo los mejores resultados. Esto resultó en que tuviera el mejor puntaje general de los 3 modelos propuestos. Luego tuvimos a la Regresión Logística con el segundo puntaje general más alto y también tuvo puntajes más altos en todas las métricas que el último modelo: **kNN**. **kNN** tuvo el porcentaje más bajo en todas las métricas que planteamos, resultando en que su puntaje general también fuera el peor. A pesar de esto, la diferencia entre el desempeño de los tres modelos es muy similar.

Considerando todo lo anterior, nuestro ganador es aquel que haya tenido el mejor desempeño en general considerando todas las métricas. Como el promedio de todas los puntajes de las métricas fue mayor en la máquina de soporte vectorial, establecemos que el modelo **SVC** definido es el ganador.
